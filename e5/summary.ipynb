```python
import numpy as np
import pandas as pd
```

```python
class Data:
    def __init__(self, path):
        self.data = pd.read_csv(path)
        self.path = path
    
    def data_path(self):
        return str(self.path)
        
    def extract(self):
        summary = pd.DataFrame({
            'X' : [self.data['x'].mean(), self.data['x'].std(), self.data['x'].min(), self.data['x'].max()],
            'Y' : [self.data['y'].mean(), self.data['y'].std(), self.data['y'].min(), self.data['y'].max()]
        },index=['mean', 'std', 'min', 'max'])
        return summary
    
    def corrco(self):
        return np.corrcoef(self.data['x'], self.data['y'])
        
    def plott(self):
        return self.data.plot()
```

```python
#reading the data and appending to list
List = []
data1 = Data('data-1.csv')
List.append(data1)
data2 = Data('data-2.csv')
List.append(data2)
data3 = Data('data-3.csv')
List.append(data3)
data4 = Data('data-4.csv')
List.append(data4)
data5 = Data('data-5.csv')
List.append(data5)
data6 = Data('data-6.csv')
List.append(data6)
```

```python
for data in List:
    print(data.data_path())
    print(data.extract())
    print('correlation coefficient between x and y: ')
    print(data.corrco()[0][1])
    #data.plott() #used for observing the relationship between x and y
    print('\n')
```
